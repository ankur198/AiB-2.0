```python
from llama_index.readers.wikipedia import WikipediaReader

loader = WikipediaReader()
documents = loader.load_data(pages=['Berlin', 'Rome', 'Tokyo', 'Canberra', 'Santiago'])
```

```python
documents
```

```python
import dotenv

dotenv.load_dotenv()
```

```python
from llama_index.llms.together import TogetherLLM

llm = TogetherLLM(model="meta-llama/Meta-Llama-3-70B-Instruct-Turbo", max_tokens=100)
```

```python
from llama_index.core import VectorStoreIndex
from llama_index.embeddings.together import TogetherEmbedding
embed_model = TogetherEmbedding(model_name="togethercomputer/m2-bert-80M-8k-retrieval")
# build an index over these Document objects.
index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)
```

```python
# you can query an index with the default QueryEngine
query_engine = index.as_query_engine(llm=llm)
response = query_engine.query("Popular places in Berlin")
print(response)
```
